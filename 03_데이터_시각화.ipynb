# matplotlib 을 이용한 단일 차트 그리기

* library 호출

```python
import matplotlib.pyplot as plt
```

* 문제: x가 1,2,3이고, y가 4,5,6인 그래프를 그리시오.

```python
x = [1,2,3]
y = [4,5,6]
```

```python
plt.figure()
plt.plot(x,y)
```

또는

```python
plt.plot(x,y)
```

- lineplot은 y값만 있으면 그림을 그릴 수 있음
- 이 때, x의 값은 0,1,2... 순서로 자동 지정됨

```python
plt.plot(y)
```

* linewidth를 이용한 선 굵기 조정

```python
plt.plot(x,y, linewidth = 10)
```

* color를 통한 선 색상 조정

```python
plt.plot(x,y, color= 'red')
```

* maker를 이용한 데이터 위치 표시

```python
plt.plot(x,y, marker='o')
```

* marker 크기 변경

```python
plt.plot(x, y, marker='o', markersize=10)
```

* 선 형태 변경

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, 'o')
```

```python
plt.plot(x, y, 'ro')
```

* 그래프 명

```python
plt.plot(x, y)
plt.title('title')
```

* 그래프 명 폰트 크기 변경

```python
plt.plot(x, y)
plt.title('title', fontsize=20)
```

* x축 명, y축 명

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')

# 크기 바꿀 거면 fontsize option넣기 
```

* 그릳그 설정 (전체)

```python
plt.plot(x, y)
plt.grid(True)

# x축만       
```
